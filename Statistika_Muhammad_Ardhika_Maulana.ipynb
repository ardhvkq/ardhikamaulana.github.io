```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Data
aplikasi = ['Instagram', 'WhatsApp', 'Facebook', 'Youtube']
penggunaan_cpu = [25, 27, 30, 10]

# Membuat Diagram Batang
plt.bar(aplikasi, penggunaan_cpu)
plt.xlabel('Aplikasi')
plt.ylabel('Range Umur')
plt.title('Rata-Rata Usia Pengguna Berbagai Aplikasi')
plt.show()

# Data
labels = ['Samsung', 'Iphone', 'Oppo', 'Vivo']
sizes = [50, 30, 15, 5]

# Membuat Diagram Lingkar
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribusi Penggunaan Sistem Operasi')
plt.show()

# Data
jam = ['10:00', '11:00', '12:00', '13:00', '14:00', '15:00', '16:00', '17:00', '18:00', '19:00']
bandwidth = [50, 65, 70, 80, 90, 85, 95, 100, 110, 120]

# Membuat Diagram Garis
plt.plot(jam, bandwidth, marker='o')  # Menambahkan penanda pada titik data
plt.xlabel('Jam')
plt.ylabel('Penggunaan Bandwidth (GB)')
plt.title('Penggunaan Bandwidth Server')
plt.grid(True)
plt.xticks(rotation=45)  # Memutar label jam agar tidak bertumpuk
plt.show()

# Data waktu respons (dalam milidetik)
response_times = [120, 130, 125, 145, 150, 165, 180, 190, 200, 210, 220, 230]

# Membuat Histogram
plt.hist(response_times, bins=5, edgecolor='black')
plt.xlabel('Waktu Respons (ms)')
plt.ylabel('Frekuensi')
plt.title('Distribusi Waktu Respons Server')
plt.show()

import matplotlib.pyplot as plt
import numpy as np

# Menghasilkan 20 data waktu respons yang didistribusikan secara normal
# Rata-rata 170 ms dengan standar deviasi 30 ms
np.random.seed(0)  # Agar hasilnya konsisten
response_times = np.random.normal(loc=170, scale=30, size=20)

# Membuat Histogram
plt.hist(response_times, bins=5, edgecolor='black')
plt.xlabel('Waktu Respons (ms)')
plt.ylabel('Frekuensi')
plt.title('Distribusi Waktu Respons Server')
plt.show()

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Menghasilkan 20 data waktu respons yang didistribusikan secara normal
np.random.seed(0)  # Agar hasil tetap konsisten
response_times = np.random.normal(loc=170, scale=30, size=20)

# Membuat Histogram
plt.hist(response_times, bins=5, edgecolor='black', density=True, alpha=0.6, color='blue')





# Membuat kurva distribusi normal di atas histogram
sns.kdeplot(response_times, color='red', lw=2)  # Menggunakan seaborn untuk membuat kurva
plt.xlabel('Waktu Respons (ms)')
plt.ylabel('Kepadatan Frekuensi')
plt.title('Distribusi Waktu Respons Server dengan Kurva Normal')

plt.show()

import matplotlib.pyplot as plt
import numpy as np


# Membaca file Excel dari lokasi D:/
file_path = "/content/PRAKTIK PENYAJIAN DATA (1).xlsx"
data = pd.read_excel(file_path)
data

# 1. Pie Chart: Pendidikan Pegawai
plt.figure(figsize=(7, 7))
data['Pendidikan Pegawai'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Distribusi Pendidikan Pegawai')
plt.ylabel('')  # Menghilangkan label y
plt.show()

# 2. Bar Chart: Lapangan Usaha Responden
plt.figure(figsize=(10, 6))
data['Lapangan Usaha Responden'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribusi Lapangan Usaha Responden')
plt.xlabel('Lapangan Usaha')
plt.ylabel('Jumlah Responden')
plt.xticks(rotation=45)
plt.show()

# 3. Line Chart: Kurs Dollar to IDR
plt.figure(figsize=(10, 6))
plt.plot(data['Kurs Dollar to IDR'], marker='o', linestyle='-', color='b')
plt.title('Pergerakan Kurs Dollar ke Rupiah')
plt.xlabel('Index Data')
plt.ylabel('Kurs Dollar to IDR')
plt.grid(True)
plt.show()

# 4. Histogram: Pengeluaran per Minggu
plt.figure(figsize=(10, 6))
plt.hist(data['Pengeluaran per Minggu'], bins=5, color='blue', edgecolor='black')
plt.title('Distribusi Pengeluaran per Minggu')
plt.xlabel('Pengeluaran (IDR)')
plt.ylabel('Frekuensi')
plt.show()

# 5. Scatter Plot: Jumlah User vs Waktu yang dihabiskan
plt.figure(figsize=(10, 6))
plt.scatter(data['Users Active'], data['Time Spent'], color='purple')
plt.title('Scatter Plot Jumlah User vs Waktu yang Dihabiskan')
plt.xlabel('Users Active')
plt.ylabel('Time Spent (minutes)')
plt.grid(True)
plt.show()

# Membuat box plot untuk Kurs Dollar to IDR
plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Kurs Dollar to IDR'])
plt.title('Box Plot Kurs Dollar to IDR')
plt.ylabel('Kurs Dollar to IDR')
plt.grid(axis='y')
plt.show()

# Membuat box plot untuk Pengeluaran per Minggu
plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Pengeluaran per Minggu'])
plt.title('Box Plot Pengeluaran per Minggu')
plt.ylabel('Pengeluaran per Minggu (IDR)')
plt.grid(axis='y')
plt.show()

# Membuat box plot untuk Users Active
plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Users Active'])
plt.title('Box Plot Users Active')
plt.ylabel('Users Active')
plt.grid(axis='y')
plt.show()


# Membuat box plot untuk Time Spent
plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Time Spent'])
plt.title('Box Plot Time Spent')
plt.ylabel('Time Spent (menit)')
plt.grid(axis='y')
plt.show()

# Misalkan data sudah dimuat ke dalam DataFrame 'data'
# Membuat box plot untuk Tinggi Badan
plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Tinggi Badan'])
plt.title('Box Plot Tinggi Badan')
plt.ylabel('Tinggi Badan (cm)')
plt.grid(axis='y')
plt.show()
```
